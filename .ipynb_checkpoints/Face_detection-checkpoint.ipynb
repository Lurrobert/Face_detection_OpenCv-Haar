```python
import cv2
import matplotlib.pyplot as plt
```

```python
classifier = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
```

```python
image = cv2.imread('test1.jpg')
```

<img src = 'test1.jpg'>

```python
boxes = classifier.detectMultiScale(image)
```

```python
print(boxes) # means that 2 bounding boxes
```

# Drawing

```python
image_detected = image.copy()
```

```python
for box in boxes:
    x1, y1, width, height = box
    x2 = x1 + width
    y2 = y1 + height
    cv2.rectangle(image_detected, (x1, y1), (x2,y2), (0,255,0), )
```

```python
plt.imshow(cv2.cvtColor(image_detected, cv2.COLOR_BGR2RGB))
```

```python
from mtcnn.mtcnn import MTCNN
```

```python
image = plt.imread('test1.jpg')
```

```python
detector = MTCNN()
```

```python
boxes = detector.detect_faces(image)
```

```python
def draw_image_with_boxes(image, boxes):
    fig, ax = plt.subplots(1)
    plt.imshow(image)
    
    for box in boxes:
        x, y, width, height = box['box']
        rect = patches.Rectangle((x, y), width, height, fill=False, color='red')
        ax.add_patch(rect)
        
        for key, value in box['keypoints'].items():
            # create and draw dot
            dot = patches.Circle(value, radius=2, color='red')
            ax.add_patch(dot)
        
    plt.show()
```

```python
draw_image_with_boxes(image, boxes)
```

```python
# 2 image
image2 = plt.imread('test2.jpg')
boxes2 = detector.detect_faces(image2)
```

```python
draw_image_with_boxes(image2, boxes2)
```

# Draw faces

```python
def draw_faces(image, result_list):
    plt.figure(figsize=(20,20))
    for i in range(len(result_list)):
        # get coordinates
        x1, y1, width, height = result_list[i]['box']
        x2, y2 = x1 + width, y1 + height
        # define subplot
        plt.subplot(1, len(result_list), i+1)
        plt.axis('off')
        # plot face
        plt.imshow(image[y1:y2, x1:x2], )
    # show the plot
    plt.show()
```

```python
plt.figure(figsize=(20,20))
```

```python
draw_faces(image2, boxes2)
```
